```python
# Fonction pour récupérer les noms des livres
def get_book_names(df4):
    return list(df4["Name"])

# Fonction pour récupérer les auteurs des livres
def get_book_authors(df4):
    return list(df4["Author"])

# Fonction pour récupérer les notes des livres
def get_book_ratings(df4):
    return list(df4["User Rating"])

# Fonction pour récupérer le nombre de commentaires pour chaque livre
def get_book_reviews(df4):
    return list(df4["Reviews"])

# Fonction pour récupérer les prix des livres
def get_book_prices(df4):
    return list(df4["Price"])

# Fonction pour récupérer les années de publication des livres
def get_book_years(df4):
    return list(df4["Year"])

# Fonction pour récupérer les genres des livres
def get_book_genres(df4):
    return list(df4["Genre"])

# Création d'un tableau contenant les données de chaque colonne
tableau = {
    "Names": get_book_names(df),
    "Authors": get_book_authors(df),
    "Ratings": get_book_ratings(df),
    "Reviews": get_book_reviews(df),
    "Prices": get_book_prices(df),
    "Years": get_book_years(df),
    "Genres": get_book_genres(df)
}

# Affichage du tableau
print(tableau)
```

```python
import requests
import pandas as pd

# Fonction pour récupérer les informations avec l'API Google Books
def get_book_info(isbn):
    url = f'https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if 'items' in data and data['items']:
            book = data['items'][0]['volumeInfo']
            if 'pageCount' in book:
                page_count = book['pageCount']
            else:
                page_count = 'N/A'
            if 'categories' in book:
                genre = book['categories'][0]
            else:
                genre = 'N/A'
            return page_count, genre
    return 'N/A', 'N/A'

# Ajout des colonnes pour le nombre de page et le genre
merged_df2['Page Count'], merged_df2['Genre'] = zip(*merged_df2['ISBN'].apply(get_book_info))

# Enregistrement du nouveau dataset
merged_df2.to_csv('finalcc.csv', index=False)
```
