```python
# Fonction pour récupérer les noms des livres
def get_book_names(df4):
    return list(df4["Name"])

# Fonction pour récupérer les auteurs des livres
def get_book_authors(df4):
    return list(df4["Author"])

# Fonction pour récupérer les notes des livres
def get_book_ratings(df4):
    return list(df4["User Rating"])

# Fonction pour récupérer le nombre de commentaires pour chaque livre
def get_book_reviews(df4):
    return list(df4["Reviews"])

# Fonction pour récupérer les prix des livres
def get_book_prices(df4):
    return list(df4["Price"])

# Fonction pour récupérer les années de publication des livres
def get_book_years(df4):
    return list(df4["Year"])

# Fonction pour récupérer les genres des livres
def get_book_genres(df4):
    return list(df4["Genre"])

# Création d'un tableau contenant les données de chaque colonne
tableau = {
    "Names": get_book_names(df),
    "Authors": get_book_authors(df),
    "Ratings": get_book_ratings(df),
    "Reviews": get_book_reviews(df),
    "Prices": get_book_prices(df),
    "Years": get_book_years(df),
    "Genres": get_book_genres(df)
}

# Affichage du tableau
print(tableau)
```

```python
import requests

def get_isbn(name, author):
    query = name + ' ' + author
    url = f'http://openlibrary.org/search.json?q={query}&limit=1'
    response = requests.get(url)
    df_final2 = response.json()
    if df_final2['num_found'] > 0:
        works = df_final2['docs'][0]['title']
        url = f'https://openlibrary.org{works["key"]}.json'
        response = requests.get(url)
        df_final2 = response.json()
        if 'isbn_13' in df_final2:
            return df_final2['isbn_13'][0]
        elif 'isbn_10' in df_final2:
            return df_final2['isbn_10'][0]
    return None
```

```python
import requests
import pandas as pd

# Fonction pour récupérer le genre d'un livre à partir de l'API Google Books
def get_book_genre(isbn):
    url = f'https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}'
    response = requests.get(url)
    data = response.json()
    try:
        genre = data['items'][0]['volumeInfo']['categories'][0]
    except:
        genre = None
    return genre

# Ajout d'une colonne de genre en utilisant l'ISBN pour récupérer les données de l'API Google Books
df_final2['Genre'] = df_final2['ISBN'].apply(get_book_genre)

# Affichage du dataframe avec la colonne de genre
print(df_final2)
```

```python
import requests
import pandas as pd

# Fonction pour récupérer l'année de publication d'un livre à partir de l'API Google Books
def get_book_year(isbn):
    url = f'https://www.googleapis.com/books/v1/volumes?q=isbn:{isbn}'
    response = requests.get(url)
    data = response.json()
    try:
        year = data['items'][0]['volumeInfo']['publishedDate'][:4]
    except:
        year = None
    return year

# Ajout d'une colonne d'année de publication en utilisant l'ISBN pour récupérer les données de l'API Google Books
df_final2['PublicationYear'] = df_final2['ISBN'].apply(get_book_year)

# Affichage du dataframe avec la colonne d'année de publication
print(df_final2)
```
